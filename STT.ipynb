```python
from logging import NullHandler
import pandas as pd

Name = "Olivia Brown"

#Questions

Q1 = "What is the Model name?"
Q2 = "When Did it happen?"
Q3 = "What is the price of your machine?"


def GetInsuarancePersonDetails(name):
  insuranceFile_path = "/content/drive/MyDrive/insurance_data.csv"
  data = pd.read_csv(insuranceFile_path)
  personDetails = data[data['Name'] == name]

  startDate = personDetails["Insurance Start Date"].values[0]
  endDate = personDetails["Insurance End Date"].values[0]
  historyOfClaims = personDetails["Claims Made"].values[0]
  modelInsuared = personDetails["Machine Insured"].values[0]
  return [startDate, endDate, historyOfClaims, modelInsuared]

def GetMachineDetails(machineName):
  washingMachineFile_path = "/content/drive/MyDrive/washing_machine.csv"
  data = pd.read_csv(washingMachineFile_path)
  machineDetails = data[data['Model Name'] == machineName]

  price = machineDetails["Price (USD)"].values[0]
  modelReleaseDate = machineDetails["Release Date"].values[0]
  review = machineDetails["Review Score (out of 5)"].values[0]
  return [price, modelReleaseDate, review]
```

```python
import torch
from transformers import AutoModelForQuestionAnswering, AutoTokenizer

model_name = "bert-large-uncased-whole-word-masking-finetuned-squad"
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# 🔹 Function to Extract Answer for Each Question
def answer_question(question, context):
    inputs = tokenizer(question, context, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)

    answer_start = torch.argmax(outputs.start_logits)
    answer_end = torch.argmax(outputs.end_logits) + 1
    answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(inputs["input_ids"][0][answer_start:answer_end]))
    return answer.strip()
```

```python


# 🔹 Run Each Question Separately and Store Answers in Variables
machine_model = answer_question(Q1, transcript)
claimDate = answer_question(Q2, transcript)
claimedPrice = answer_question(Q3, transcript)

insuaredData = GetInsuarancePersonDetails(Name)
machineData = GetMachineDetails(machine_model.upper())
```

```python
import re

def extract_full_number(text):
    """Extracts a number from text, handling spaces and commas, and converts it to an integer."""
    match = re.search(r'[\d,\s]+', text)  # Find numbers with spaces and commas
    if match:
        cleaned_number = match.group().replace(',', '').replace(' ', '')  # Remove spaces and commas
        return int(cleaned_number)  # Convert to integer
    return None

claimedPrice = extract_full_number(claimedPrice)
```

```python
def percentage_difference(original, claimed):
    """Calculates percentage difference between two numbers."""
    if original == 0:
        return "Undefined (Original value is 0)"
    difference = abs(claimed - original)
    percent_diff = (difference / original) * 100
    return round(percent_diff, 2)  # Rounded to 2 decimal places
```
